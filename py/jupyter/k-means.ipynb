```python
import pandas as pd 
df =pd.read_csv('datasets/income.csv')
df.head()
```

```python
from matplotlib import pyplot as plt 
# %matplotlib inline
plt.scatter(df['Age'],df['Income($)'])
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
km=KMeans(n_clusters=3)
km
```

```python
y_pred=km.fit_predict(df[['Age','Income($)']])
y_pred
```

```python
df['cluster'] = y_pred
df.head()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]

plt.scatter(df1['Age'],df1['Income($)'],color='green')
plt.scatter(df2['Age'],df2['Income($)'],color='red')
plt.scatter(df3['Age'],df3['Income($)'],color='black')

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*') #centroid

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
scaler = MinMaxScaler()

scaler.fit(df[['Income($)']])
df['Income($)']= scaler.transform(df[['Income($)']])

scaler.fit(df[['Age']])
df['Age']= scaler.transform(df[['Age']])

df
```

```python
y_pred1=km.fit_predict(df[['Age','Income($)']])
y_pred1
```

```python
df['cluster1'] = y_pred1
df.head()
```

```python
km.cluster_centers_
```

```python
df4=df[df.cluster1==0]
df5=df[df.cluster1==1]
df6=df[df.cluster1==2]

plt.scatter(df4['Age'],df4['Income($)'],color='green')
plt.scatter(df5['Age'],df5['Income($)'],color='red')
plt.scatter(df6['Age'],df6['Income($)'],color='black')

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*') #centroid

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.legend()
```

```python
# Elbow method
k_rng = range(1,10)
sse=[]
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)
    
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(k_rng,sse)
```
