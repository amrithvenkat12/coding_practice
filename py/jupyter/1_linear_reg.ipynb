```python
import pandas as pd 
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/homeprices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df['price'])
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg.predict([[3300]])#regressor in python3+ always 2Darray
```

```python
d=pd.read_csv('datasets/areas.csv')
d.head()
```

```python
p=reg.predict(d)
p
```

```python
d['pred']=p
d
```

```python
d.to_csv('pred.csv',index=False)
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```
