```python
dummies = pd.get_dummies(df['Car Model'])
dummies
```

```python
final=pd.concat([df,dummies],axis='columns').drop(['Car Model','Mercedez Benz C class'],axis='columns')
final
```

```python
X=final.drop(['Sell Price($)'],axis='columns')
X
```

```python
y=final['Sell Price($)']
y
```

```python
# #Linear Regression
from sklearn.linear_model import LinearRegression
model=LinearRegression()

model.fit(X,y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```

```python
model.score(X,y)
```
